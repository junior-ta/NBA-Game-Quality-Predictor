```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
from datetime import datetime, timedelta
```

```python
#Creatimg my dataframe with column names scraped beforehand
df= pd.DataFrame(columns=['Rank',
 'Season',
 'Date',
 'Score',
 'Excitment',
 'Tension',
 'Comeback',
 'MVP',
 'LVP'])

start_date = datetime(2024, 10, 22)
end_date = datetime(2025, 4, 13)

current_date = start_date
while current_date <= end_date:
    #print(current_date.strftime("%Y-%m-%d"))  # or just use current_date
    url= f"https://stats.inpredictable.com/nba/topGame.php?sort=ei&season=2024&team=ALL&po=0&frdt={current_date.strftime('%Y-%m-%d')}&todt={current_date.strftime('%Y-%m-%d')}"
    page= requests.get(url)
    soup= BeautifulSoup(page.text, 'html')

    table= soup.find('table')

    ##Getting columns
    columns= soup.find_all('th')
    columns_list= [column.text for column in columns]

    #getting rows
    rows= table.find_all('tr')[1:]

    rows_list=[]
    for row in rows:
        curr= row.find_all('td')
        curr.pop(4)
        rows_list.append([data.text for data in curr])

    for row in rows_list:
        df.loc[len(df)]= row

    current_date += timedelta(days=1)

df.drop(['Rank'], axis=1, inplace=True)
df.to_csv()
df
```
